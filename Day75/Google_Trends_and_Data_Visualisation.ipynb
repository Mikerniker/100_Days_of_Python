### Notebook Formatting & Style Helpers

```python
# Register date converters to avoid warning messages
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# Create locators for ticks on the time axis
years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y')
```

```python
# Regenerated line graph with formatted ticks

# increase size and resolution
plt.figure(figsize=(14,8), dpi=120)

# display the title
plt.title("Tesla Web Search vs Price", fontsize=18)

# Increase the size and rotate the labels on the x-axis
plt.xticks(fontsize=14, rotation=45)

ax1 = plt.gca() # get the axis
ax2 = ax1.twinx() # create another axis that shares the same x-axis

# Increase fontsize and linewidth for larger charts
ax1.set_xlabel('Year', fontsize=14)
ax1.set_ylabel('TSLA Stock Price', color="#FB6736", fontsize=14)    # HEX code
ax2.set_ylabel('Search Trend', color='mediumslateblue', fontsize=14) # named colour

# Set the minimum and maximum values on the axes
ax1.set_ylim([0, 600])
ax1.set_xlim([df_tesla.MONTH.min(), df_tesla.MONTH.max()])

#Format the ticks
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

ax1.plot(df_tesla.MONTH, df_tesla.TSLA_USD_CLOSE, color="#FB6736", linewidth=3)
ax2.plot(df_tesla.MONTH, df_tesla.TSLA_WEB_SEARCH, color='mediumslateblue', linewidth=3)

# Display chart explicitly
plt.show()
```
