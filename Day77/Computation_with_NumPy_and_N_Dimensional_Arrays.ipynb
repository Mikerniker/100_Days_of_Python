# Understanding NumPy's ndarray

NumPy's most amazing feature is the **powerful** ndarray.

<img src="https://i.imgur.com/1IUqnxX.png" width=200>

#### 1-Dimensional Arrays (Vectors)

```
# 1-dimensional array (i.e., a “vector”)

my_array = np.array([1.1, 9.2, 8.1, 4.7])
```

```
# Show rows and columns

my_array.shape
```

```
# Access an element in a ndarray by the element's index, similar to a Python List:

my_array[2]
```

```
# Show dimensions of an array
my_array.ndim
```

#### 2-Dimensional Arrays (Matrices)

```
#  2-dimensional array (i.e., a “matrix”)

array_2d = np.array([[1, 2, 3, 9],
                     [5, 6, 7, 8]])
```

```
# NumPy refers to the dimensions as axes,
# so the first axis has length 2 and the second axis has length 4.
print(f'array_2d has {array_2d.ndim} dimensions')
print(f'Its shape is {array_2d.shape}')
print(f'It has {array_2d.shape[0]} rows and {array_2d.shape[1]} columns')
print(array_2d)
```

```
# To access a particular value, provide an index for each dimension.
# For two dimensions, provide an index for the row and for the column.
# Example: To access the 3rd value in the 2nd row:

array_2d[1,2]
```

```
# To access an entire row and all the values therein, use the : operator
# Like you would do with a Python List. Example for the entire first row:

array_2d[0, :]
```

# Broadcasting and Scalars


```
# Broadcasting
# In mathematics, a single number is often called a scalar.
array_2d = np.array([[1, 2, 3, 4],
                     [5, 6, 7, 8]])
```

```
# If you want to add a scalar to an array
# Broadcasting makes the scalar compatible with the larger array.

array_2d + 10
```

```
# Also works with multiplication

array_2d * 5
```

# Matrix Multiplication with @ and .matmul()

<img src=https://i.imgur.com/LCdhmi8.png width=350>

If we're not multiplying our ndarray by a single number, ex if we multiply it by another vector or a 2-dimensional array, we follow the rules of linear algebra.

```
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])

b1 = np.array([[4, 1, 3],
               [5, 8, 5]])

print(f'{a1.shape}: a has {a1.shape[0]} rows and {a1.shape[1]} columns.')
print(f'{b1.shape}: b has {b1.shape[0]} rows and {b1.shape[1]} columns.')
print('Dimensions of result: (4x2)*(2x3)=(4x3)')
```

**Challenge**: Convert the image to black and white. The values in our `img` range from 0 to 255.
* Divide all the values by 255 to convert them to sRGB, where all the values are between 0 and 1.
* Next, multiply the sRGB array by the `grey_vals` to convert the image to grey scale.
* Finally use Matplotlib's [`.imshow()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.imshow.html) together with the colormap parameter set to gray `cmap=gray` to look at the results.

```
# Divide all the values by 255 to convert them to sRGB
sRGB_array = img / 255
```

```
grey_vals = np.array([0.2126, 0.7152, 0.0722])
```

```
#  Multiply the sRGB array by the grey_vals to convert the image to grey scale.
# Can either multiply them together with the @ operator or the .matmul() function.
# Alternative img_gray = np.matmul(sRGB_array, grey_vals)

img_gray = sRGB_array @ grey_vals
```

```
# Show the image with Matplotlib
# The cmap parameter is important here because
# If we leave it out the function will not know that is dealing with a black and white image.

plt.imshow(img_gray, cmap='gray')
```

#### Challenge Solutions

```
# 1. Flip the grayscale image upside down
# Alternative: flipped_raccoon = np.flipud(img_gray)
# plt.imshow(flipped_raccoon, cmap='gray')

plt.imshow(np.flip(img_gray), cmap='gray')
```

```
# 2. Rotate the colour image
# Shorter version: plt.imshow(np.rot90(img))

rotated_raccoon = np.rot90(img)
plt.imshow(rotated_raccoon)
```

```
# 3. Invert (i.e., solarize) the colour image.
# To do this you need to converting all the pixels to their "opposite" value, so black (0) becomes white (255).
# Use of NumPy's broadcast when doing operations with a scalar.
# In this case, scalar is 255 - the maximum value for a pixel in RGB
# If we subtract the values of img from 255, then we get the opposite value for each pixel:

solar_img = 255 - img
plt.imshow(solar_img)
```
