```python
import pandas as pd
```

```python
df = pd.read_csv('salaries2023.csv')
```

```python
df.isna()
```

```python
df.tail()
```

```python
# Make new column to clean data for Early Career
clean_early_career = df['Early Career Pay'].str.replace('$', '')
df.insert(1, 'Early Career New', clean_early_career)
df.head()
```

```python
df['Early Career New'].str.replace('"', '')
df.head()
```

```python
# Make new column for Mid-Career Pay
clean_mid_career = df['Mid-Career Pay'].str.replace('$', '')
df.insert(1, 'Mid-Career New', clean_mid_career)
df['Mid-Career New'].str.replace('"', '')
df.head()
```

```python
# Added this to remove the commas since I was getting a TypeError
df['Mid-Career New'] = df['Mid-Career New'].str.replace(',', '').astype(float)
```

```python
df['Early Career New'] = df['Early Career New'].str.replace(',', '').astype(float)
```
